```python
# Gerekli Kütüphaneler
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

# 1. Veri Seti Kaynağı: UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00451/dataR2.csv"

# UCI'den veri setini çekme
print("UCI Machine Learning Repository'den veri seti çekiliyor...")
lung_cancer_data = pd.read_csv(url)

# İlk birkaç satırı kontrol etme
print("\nVeri setinin ilk 5 satırı:")
print(lung_cancer_data.head())

# 2. Özellikler ve Etiketlerin Ayrılması
features = lung_cancer_data.iloc[:, :-1].values  # Özellik sütunları (son sütun hariç)
labels = lung_cancer_data.iloc[:, -1].values    # Etiket sütunu (son sütun)

# 3. Eğitim ve Test Verilerini Bölme
train, test, train_labels, test_labels = train_test_split(features, labels, test_size=0.2, random_state=42)

# 4. Gaussian Naive Bayes ile Sınıflandırma
print("\nGaussian Naive Bayes ile Model Eğitimi:")
nb_model = GaussianNB()  # Gaussian Naive Bayes modeli oluşturma
nb_model.fit(train, train_labels)  # Modeli eğitme
nb_predictions = nb_model.predict(test)  # Test verilerinde tahmin yapma
nb_accuracy = accuracy_score(test_labels, nb_predictions)  # Doğruluk hesaplama
print(f"Naive Bayes Doğruluğu: {nb_accuracy:.2f}")

# 5. Linear Regression ile Model Eğitimi
print("\nLinear Regression ile Model Eğitimi:")
lr_model = LinearRegression()  # Lineer regresyon modeli oluşturma
lr_model.fit(train, train_labels)  # Modeli eğitme
lr_predictions = lr_model.predict(test)  # Test verilerinde tahmin yapma
print("İlk 5 Gerçek ve Tahmin Edilen Etiketler:")
print(f"Gerçek Etiketler: {test_labels[:5]}")  # İlk 5 gerçek etiket
print(f"Tahmin Edilen Etiketler: {lr_predictions[:5]}")  # İlk 5 tahmin edilen değer
```

```python
from IPython import get_ipython
from IPython.display import display
# %%
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
# %%
data = load_breast_cancer()
label_names = data["target_names"]
labels = data["target"]
feature_names = data["feature_names"]
features = data["data"]
# %%
print(label_names)
print("Class label :", labels[0])
print(feature_names)
print(features[0], "\n")
# %%
train, test, train_labels, test_labels = train_test_split(features, labels,
                                                          test_size=0.2,
                                                          random_state=42)
# %%
nb = GaussianNB()
nb.fit(train, train_labels) # Changed gnb to nb
# %%
# Predict on the test data using the trained model
preds = nb.predict(test)
# Now, preds holds the model's predictions
print(accuracy_score(test_labels, preds))
```
